```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
data = pd.read_csv('household_income_20221130_MS.csv', delimiter=',', quotechar='"')

# Filter rows where occupation is farming
filtered_data = data[data['STATE'].isin(['Karnataka'])]
filtered_data = data[data['DISTRICT'].isin(['Bellary', 'Davanagere','Chitradurga', 'Haveri', 'Gadag'])]
filtered_data = data[data['OCCUPATION_GROUP'].isin(['Organised Farmers', 'Small/Marginal Farmers', 'Agricultural Labourers'])]

# Create a new DataFrame with required columns
new_data = filtered_data[['HH_ID', 'OCCUPATION_GROUP', 'EDUCATION_GROUP', 'TOTAL_INCOME']]

#  Save the filtered data to a new CSV file
new_data.to_csv('filtered_data.csv', index=False)
```

```python

# Step 5: Visualize data using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=new_data, x='EDUCATION_GROUP')
plt.title('Education Distribution by Occupation')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```
